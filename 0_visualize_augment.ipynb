```python
from tensorflow import keras
import tensorflow as tf
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
```

#### ทำการ pre-processing โดยใช้ binary-threshold เพื่อเร่งค่าสีให้มีความชัดเจนมากยิ่งขึ้นในกรณีที่ภาพ datasets หรือ input data ภาพมีความจาง

```python
#https://www.geeksforgeeks.org/python-thresholding-techniques-using-opencv-set-1-simple-thresholding/ 
      
# organizing imports  
import cv2  
import numpy as np  
image1 = cv2.imread(img_path)  
  
img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY) 


ret, thresh1 = cv2.threshold(img, 255-0.1, 255, cv2.THRESH_BINARY) 

  
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(image1, 'gray')

plt.subplot(1, 2, 2)
plt.imshow(thresh1, 'gray')
cv2.imwrite('./a.png', thresh1)
```

```python
img_path = './a.png'
```

#### input image ที่ได้หลังจากการทำ binary-threshold

```python
#Loads image in from the set image path
img = keras.preprocessing.image.load_img(img_path, target_size= (100,100))
img_tensor = keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
#Allows us to properly visualize our image by rescaling values in array
#Plots image
img_tensor /= 255.
plt.figure(figsize=(6,6))
plt.imshow(img_tensor[0])
plt.show()
```

#### augment image โดยการ shift width กับ height  เป็นหลักเนื่องจากข้อมูลเป็นตัวอักษรจึงไม่ควรทำการ rotation หรือ crop ภาพ



```python
#Loads in image path
img = keras.preprocessing.image.load_img(img_path, target_size= (100,100))
img_tensor = keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
#Uses ImageDataGenerator to flip the images

datagen = ImageDataGenerator(height_shift_range=0.1,
                             width_shift_range=0.1,
                             #zoom_range=0.3,
                             
                             
                            )


#Creates our batch of one image
pic = datagen.flow(img_tensor, batch_size = 1)
plt.figure(figsize=(16,16))
#Plots our figures
for i in range(1,17):
  plt.subplot(4, 4, i)
  batch = pic.next()
  image_ = batch[0].astype('uint8')
  plt.imshow(image_, 'gray')
plt.show()
```

```python
#Loads image in from the set image path
img = keras.preprocessing.image.load_img(img_path, target_size= (100,100), color_mode='grayscale')
img_tensor = keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
#Allows us to properly visualize our image by rescaling values in array
#Plots image
img_tensor /= 255.
plt.figure(figsize=(6,6))
plt.imshow(img_tensor[0], 'gray')
plt.show()
```
